# Label Encoding

##### In machine learning, we usually deal with datasets that contain multiple labels in one or more than one columns. These labels can be in the form of words or numbers. To make the data understandable or in human-readable form, the training data is often labelled in words 

```python
from sklearn.preprocessing import LabelEncoder
label= LabelEncoder()
data.loc[:,'Attrition']= label.fit_transform(data.loc[:,'Attrition'])
data.loc[:,'BusinessTravel']= label.fit_transform(data.loc[:,'BusinessTravel'])
data.loc[:,'Department']= label.fit_transform(data.loc[:,'Department'])
data.loc[:,'EducationField']= label.fit_transform(data.loc[:,'EducationField'])
data.loc[:,'Gender']= label.fit_transform(data.loc[:,'Gender'])
data.loc[:,'JobRole']= label.fit_transform(data.loc[:,'JobRole'])
data.loc[:,'MaritalStatus']= label.fit_transform(data.loc[:,'MaritalStatus'])
data.loc[:,'Over18']= label.fit_transform(data.loc[:,'Over18'])
data.loc[:,'OverTime']= label.fit_transform(data.loc[:,'OverTime'])
```

# Train -Test Split

```python
x=data.drop('Attrition', axis=1)
y=data['Attrition']
```

```python
y.value_counts()
```

# Data is Imbalanced
##### Over sampling technique is used

```python
from imblearn.over_sampling import RandomOverSampler
rvs = RandomOverSampler(random_state=42)
```

```python
x_rvs, y_rvs=rvs.fit_resample(x,y)
```

```python
y_rvs.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x_rvs, y_rvs, random_state=42, test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, precision_score, recall_score
```

```python
logreg= LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
log_pred = logreg.predict(x_test)
```

```python
accuracy_score(y_test, log_pred)
```

```python
precision_score(y_test, log_pred)
```

```python
recall_score(y_test, log_pred)
```

```python
confusion_matrix(y_test, log_pred)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

#### GridSearchCV is used to find out best hyperparameter.

```python
param_grid={'max_depth':[2,3,4,5],
           'max_features':[1,3,5,10],
           'min_samples_split':[2,3,5,10],
           'criterion':['gini','entropy'],
           'min_samples_leaf':[1,2,3,4]
           }
# Make the object of the class GridSearchCV
grid=GridSearchCV(dtree,param_grid,verbose=2)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
dtree1= DecisionTreeClassifier(criterion= 'gini',
                               max_depth= 5,
                               max_features= 10,
                               min_samples_leaf= 4,
                               min_samples_split= 2)
```

```python
dtree1.fit(x_train, y_train)
```

```python
dtree_pred=dtree1.predict(x_test) 
```

```python
confusion_matrix(y_test, dtree_pred)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=20, criterion= 'gini',
                               max_depth= 5,
                               max_features= 10,
                               min_samples_leaf= 4,
                               min_samples_split= 2)
```

```python
forest.fit(x_train, y_train)
```

```python
forest_pred= forest.predict(x_test)
```

```python
accuracy_score(y_test, forest_pred)
```

```python
precision_score(y_test, forest_pred)
```

```python
recall_score(y_test, forest_pred)
```

```python
confusion_matrix(y_test, forest_pred)
```

```python
final_data=pd.DataFrame({'Models':['Logistic_regression', 'Decision_tree','Random_forest'],
             'Accuracy':[accuracy_score(y_test, log_pred)*100,
                        accuracy_score(y_test, dtree_pred)*100,
                        accuracy_score(y_test, forest_pred)*100],
             'Precision':[precision_score(y_test, log_pred)*100,
                         precision_score(y_test, dtree_pred)*100,
                         precision_score(y_test, forest_pred)*100],
             'Recall':[recall_score(y_test, log_pred)*100,
                      recall_score(y_test, dtree_pred)*100,
                      recall_score(y_test, forest_pred)*100]}, index=(1,2,3))
```
